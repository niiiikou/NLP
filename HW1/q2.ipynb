# 1. White Space Tokenization

## for english book

```python
en_sentences = nltk.sent_tokenize(en_text)
```

```python
en_sentences[0]
```

```python
from nltk.tokenize import WhitespaceTokenizer
```

```python
en_WS_tokens = []
for sen in en_sentences:
  en_WS_tokens += WhitespaceTokenizer().tokenize(sen)
```

```python
len(en_WS_tokens)
```

```python
print(WhitespaceTokenizer().tokenize(en_input))
```

## for farsi book

```python
fa_sentences = nltk.sent_tokenize(fa_text)
```

```python
fa_sentences[0]
```

```python
fa_WS_tokens = []
for sen in fa_sentences:
  fa_WS_tokens += WhitespaceTokenizer().tokenize(sen)
```

```python
len(fa_WS_tokens)
```

```python
print(WhitespaceTokenizer().tokenize(fa_input))
```

# 2. Spacy 

```python
import spacy
```

## for english book

```python
nlp = spacy.load("en_core_web_sm")
```

```python
en_SP_tokens = []
for sen in en_sentences:
  for token in nlp(sen):
    en_SP_tokens.append(token.text)
```

```python
len(en_SP_tokens)
```

```python
en_input_SP_tokens = []
for token in nlp(en_input):
  en_input_SP_tokens.append(token.text)

print(en_input_SP_tokens)
```

## for farsi book

```python
from spacy.lang.fa import Persian
nlp = Persian()
tokenizer = nlp.tokenizer
#tokens = tokenizer(fa_sentences[0])
```

```python
fa_SP_tokens = []
for sen in fa_sentences:
  for token in tokenizer(sen):
    fa_SP_tokens.append(token.text)
```

```python
len(fa_SP_tokens)
```

```python
print(fa_SP_tokens)
```

```python
fa_input_SP_tokens = []
for token in tokenizer(fa_input):
  fa_input_SP_tokens.append(token.text)

print(fa_input_SP_tokens)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

## for english book

```python
tokenizer = Tokenizer(BPE())

trainer = BpeTrainer()

tokenizer.pre_tokenizer = Whitespace()

files = ["hp_en.txt" for split in ["test", "train", "valid"]]
tokenizer.train(files, trainer)
```

```python
num_tokens = 0
for k, v in tokenizer.get_vocab().items():
  num_tokens += v
```

```python
num_tokens
```

```python
output = tokenizer.encode(en_input)
print(output.tokens)
```
